### Pregunta 3
en la primera parte se hizo con numpy en  la pregunta tres se introdujo la parte tensorial

```python
import numpy as np
import tensorflow as tf

# ### Pregunta 1
I3 = np.array([[1,0,0],[0,1,0],[0,0,1]])
# ahora extraer cada vector columna de la matriz I3
v1 = I3[:,0]
v2 = I3[:,1]
v3 = I3[:,2]

# Convertir los vectores NumPy a tensores de TensorFlow
v1_tf = tf.convert_to_tensor(v1, dtype=tf.float32)
v2_tf = tf.convert_to_tensor(v2, dtype=tf.float32)
v3_tf = tf.convert_to_tensor(v3, dtype=tf.float32)

# realizar el producto punto entre v1 y v2 usando TensorFlow
producto_ortogonal_1_2 = tf.tensordot(v1_tf, v2_tf, axes=1)
print(producto_ortogonal_1_2)

producto_ortogonal_1_3 = tf.tensordot(v1_tf, v3_tf, axes=1)
print(producto_ortogonal_1_3)

producto_ortogonal_2_3 = tf.tensordot(v2_tf, v3_tf, axes=1)
print(producto_ortogonal_2_3)


# ### Pregunta 2
# calcular la norma unitaria de v1, v2 y v3 usando TensorFlow

# Calcular la norma unitaria de v1, v2 y v3
norma_v1 = tf.norm(v1_tf)
print(f"La norma unitaria de v1 es: {norma_v1}")

norma_v2 = tf.norm(v2_tf)
print(f"La norma unitaria de v2 es: {norma_v2}")

norma_v3 = tf.norm(v3_tf)
print(f"La norma unitaria de v3 es: {norma_v3}")
```

### Problema 4

```python
K = np.array([[2/3,1/3,2/3],[-2/3,2/3,1/3],[1/3,2/3,-2/3]])
```

```python
# repetir los problemas del 1 al 3 pero ahora usando la matriz k

import numpy as np
import tensorflow as tf

# Problema 1 con la matriz K
K = np.array([[2/3,1/3,2/3],[-2/3,2/3,1/3],[1/3,2/3,-2/3]])
k1 = K[:,0]
k2 = K[:,1]
k3 = K[:,2]

producto_ortogonal_k1_k2 = np.dot(k1, k2)
print(f"Producto punto entre k1 y k2: {producto_ortogonal_k1_k2}")

producto_ortogonal_k1_k3 = np.dot(k1, k3)
print(f"Producto punto entre k1 y k3: {producto_ortogonal_k1_k3}")

producto_ortogonal_k2_k3 = np.dot(k2, k3)
print(f"Producto punto entre k2 y k3: {producto_ortogonal_k2_k3}")


# Problema 2 con la matriz K
norma_k1 = np.linalg.norm(k1)
print(f"La norma unitaria de k1 es: {norma_k1}")

norma_k2 = np.linalg.norm(k2)
print(f"La norma unitaria de k2 es: {norma_k2}")

norma_k3 = np.linalg.norm(k3)
print(f"La norma unitaria de k3 es: {norma_k3}")


# Problema 3 con la matriz K (usando TensorFlow)
k1_tf = tf.convert_to_tensor(k1, dtype=tf.float32)
k2_tf = tf.convert_to_tensor(k2, dtype=tf.float32)
k3_tf = tf.convert_to_tensor(k3, dtype=tf.float32)

producto_ortogonal_k1_k2_tf = tf.tensordot(k1_tf, k2_tf, axes=1)
print(f"Producto punto entre k1 y k2 (TensorFlow): {producto_ortogonal_k1_k2_tf}")

producto_ortogonal_k1_k3_tf = tf.tensordot(k1_tf, k3_tf, axes=1)
print(f"Producto punto entre k1 y k3 (TensorFlow): {producto_ortogonal_k1_k3_tf}")

producto_ortogonal_k2_k3_tf = tf.tensordot(k2_tf, k3_tf, axes=1)
print(f"Producto punto entre k2 y k3 (TensorFlow): {producto_ortogonal_k2_k3_tf}")

norma_k1_tf = tf.norm(k1_tf)
print(f"La norma unitaria de k1 es (TensorFlow): {norma_k1_tf}")

norma_k2_tf = tf.norm(k2_tf)
print(f"La norma unitaria de k2 es (TensorFlow): {norma_k2_tf}")

norma_k3_tf = tf.norm(k3_tf)
print(f"La norma unitaria de k3 es (TensorFlow): {norma_k3_tf}")
```
